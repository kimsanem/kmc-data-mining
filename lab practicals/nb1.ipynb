```python
df = pd.read_csv('../dataset/weather.csv')
print(df.head())
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB

# Your dataset
data = {
    "Outlook": ["Sunny", "Sunny", "Overcast", "Rain", "Rain"],
    "Temperature": ["Hot", "Hot", "Hot", "Mild", "Cool"],
    "Humidity": ["High", "High", "High", "High", "Normal"],
    "Wind": ["Weak", "Strong", "Weak", "Weak", "Weak"],
    "PlayTennis": ["No", "No", "Yes", "Yes", "Yes"]
}

df = pd.DataFrame(data)

# Encode all categorical columns
encoders = {}
for col in df.columns:
    enc = LabelEncoder()
    df[col] = enc.fit_transform(df[col])
    encoders[col] = enc

# Split features and label
X = df.drop("PlayTennis", axis=1)
y = df["PlayTennis"]

# Train Naive Bayes model
model = GaussianNB()
model.fit(X, y)

# Example prediction
test = pd.DataFrame([{
    "Outlook": "Sunny",
    "Temperature": "Cool",
    "Humidity": "High",
    "Wind": "Weak"
}])

# Encode test data with same encoders
for col in test.columns:
    test[col] = encoders[col].transform(test[col])

pred = model.predict(test)[0]
prediction_label = encoders["PlayTennis"].inverse_transform([pred])[0]

print("Predicted PlayTennis:", prediction_label)
```
