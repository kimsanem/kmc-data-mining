```python
import numpy as np
```

```python
rand_num = np.random.randint(0, 10, 10)

# for any range beside [0,1]
# max_range = int(input("Enter the maximun number for the range: "))
# min_range = int(input("Enter the minimun number for the range: "))
```

```python
def min_max(rand_num, min_range=0, max_range=1):
    min_num = min(rand_num)
    max_num = max(rand_num)
    return (min_range + ((rand_num - min_num)/max_range))/(max_num - min_num) * (max_range - min_range) 
```

```python
def z_score(rand_num):
    mean = np.mean(rand_num)
    std_dev = np.std(rand_num)
    return ((rand_num - mean) / std_dev)
```

```python
print("Original Random Numbers: ", rand_num)
print("Min-Max Normalized: ", min_max(rand_num))
print("Z-Score Standardized: ", z_score(rand_num))
```
